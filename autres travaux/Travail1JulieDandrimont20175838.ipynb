### Création du dataframe


```python
#importation des données et création du dataframe
df = pd.read_csv('insurance.csv')
df
```

**age**: age of primary beneficiary

**sex**: insurance contractor gender, female, male

**bmi**: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height,
objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9

**children**: Number of children covered by health insurance / Number of dependents

**smoker**: Smoking

**region**: the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.

**charges**: Individual medical costs billed by health insurance

###  implémentation Numpy

```python
#convertis le dataframe en numpy array et crée une nouvelle matrice
df_numpy = df.to_numpy()
BmiCharges = np.zeros((observations, 4)) 
#print(BmiCharges.shape) 


#remplis la nouvelle matrice à partir des données du numpy 
BmiCharges[:,0] = df_numpy[:,colonnes.index('bmi')]
BmiCharges[:,1] = df_numpy[:,colonnes.index('charges')]
BmiCharges[:,2] = (df_numpy[:,colonnes.index('children')])+1
# afin de calculer une nouvelle colonne
BmiCharges[:,3] = (BmiCharges[:,1]/BmiCharges[:,2])/BmiCharges[:,0]

# permet de discrétiser l'imc
for i in range(observations) :
    if BmiCharges[i,0] >= 24.9: 
        BmiCharges[i,0] = 1 # code pour le surpoids
        
    elif BmiCharges[i,0] < 18.5 :
        BmiCharges[i,0] = 2 #code pour la maigreur
        
    else :
        BmiCharges[i,0] = 3 #code pour la normalité 

#convertis la matrice en dataframe        
dfBmiCharges = pd.DataFrame(BmiCharges,columns = ['imc','charges','bénéficitaires','charge_par_gramme_de_graisse'])
dfBmiCharges
```
